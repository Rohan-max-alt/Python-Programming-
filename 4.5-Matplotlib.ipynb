## MatplotLib

Matplotlib is a plotting library package for Python and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the disadvantages of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt  # matplotlib library file 

%matplotlib inline
```

```python
import numpy as np # numpy for array or storage 
```

```python
## Simple Examples

x=np.arange(0,10)   # this gives the values from 0 to 10 
y=np.arange(11,21)  # from 11 to 21
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='g') # here x is x-axis and y is y-axis and c is the color and 'g' is green.
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png') # this function daves the figure as Test.png in the directory.

```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12) # the 'r*' is the Stars and dashes is the space linned line. 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

# Subplots are small plots of multiple grphs that fit in flot format
plt.subplot(2,2,1)
plt.plot(x,y,'r--') 
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11)  # this is the range 
y = 3 * x + 5        # y gives us the straight line 
plt.title("Matplotlib demo")  
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x)  # usully sin values and cos values are zigzag in graph from -1 to +1
y_cos = np.cos(x)  
z=range(0,159)   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210] # these are the percentage of content or values of each element. 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice # explode takes away one of the slices from pie.

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False)
# plt.pie gives us access to plot a pie chart
plt.axis('equal')
plt.show()
```
